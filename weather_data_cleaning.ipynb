```python
weatherHourly = weatherAll[weatherAll['REPORT_TYPE']=='FM-15']
weatherHourly.dropna(axis = 1, how = 'all', inplace = True)
```

```python
weatherHourly.drop(['HourlyPresentWeatherType', 'HourlyPressureChange', 'HourlyPressureTendency', 'HourlyWindGustSpeed', 
                    'HourlyWindDirection', 'REM', 'REPORT_TYPE.1', 'SOURCE.1', 'REPORT_TYPE', 'SOURCE', 'HourlyAltimeterSetting', 
                    'HourlySkyConditions', 'HourlyPrecipitation', 'HourlyDewPointTemperature', 'HourlyDryBulbTemperature', 
                    'HourlySeaLevelPressure', 'STATION'], axis=1, inplace = True)
```

```python
weatherHourly.tail()
```

```python
weatherHourly.describe()
```

set day_id, hour_id

```python
weatherHourly['DATE'] = pd.to_datetime(weatherHourly['DATE'])
```

```python
weatherHourly.tail()
```

```python
weatherHourly['Month'] = (weatherHourly['DATE']).apply(lambda x: x.month).tolist()
weatherHourly['Day'] = (weatherHourly['DATE']).apply(lambda x: x.day).tolist()
weatherHourly['Hour'] = (weatherHourly['DATE']).apply(lambda x: x.hour).tolist()
```

```python
weatherHourly.head()
```

```python
from datetime import datetime
weatherHourly['day_id'] = (weatherHourly['DATE']).apply(lambda x: (x - datetime.strptime('2018-01-01',  "%Y-%M-%d")).days).tolist()
weatherHourly['hour_id'] = (weatherHourly['DATE']).apply(lambda x: ((x - datetime.strptime('2018-01-01',  "%Y-%M-%d")).days)*24 + (x - datetime.strptime('2018-01-01',  "%Y-%M-%d")).seconds/3600).tolist()
```

```python
weatherHourly.head()
```

Fill na

before 10.13

```python
sum(weatherHourly.isna())
```

```python
weatherHourlyWithoutAbnormal = weatherHourly[weatherHourly['DATE']<'2018-10-13']
```

```python
sum(weatherHourlyWithoutAbnormal.isna())
```

```python
weatherHourlyWithoutAbnormal = weatherHourlyWithoutAbnormal.fillna(method='ffill')
```

after 10.13

```python
weatherHourlyWithAbnormal = weatherHourly[weatherHourly['DATE']>='2018-10-13']
```

```python
sum(weatherHourlyWithAbnormal.isna())
```

```python
weatherHourlyWithAbnormal = weatherHourlyWithAbnormal.fillna(method='ffill')
```

```python
weatherDaily = pd.read_csv("weather_daily.csv")
```

```python
weatherDaily = weatherDaily[['DATE', 'DailyAverageWindSpeed']]
```

```python
weatherDaily['DATE'] = pd.to_datetime(weatherDaily['DATE'])
```

```python
from datetime import datetime
weatherDaily['day_id'] = (weatherDaily['DATE']).apply(lambda x: (x - datetime.strptime('2018-01-01',  "%Y-%M-%d")).days).tolist()
```

```python
weatherDaily = weatherDaily[[ 'DailyAverageWindSpeed', 'day_id']]
```

```python
weatherDaily.head()
```

```python
weatherHourlyWithAbnormal.head()
```

```python
weatherHourlyWithAbnormal = weatherHourlyWithAbnormal.merge(weatherDaily, left_on = 'day_id', right_on = 'day_id', how = 'left')
```

```python
weatherHourlyWithAbnormal.drop(['HourlyWindSpeed'], axis = 1, inplace = True)
```

```python
weatherHourlyWithAbnormal = weatherHourlyWithAbnormal.rename(columns = {'DailyAverageWindSpeed': 'HourlyWindSpeed'})
```

```python
weatherHourlyWithAbnormal.head()
```

Get the whole year cleaned data

```python
weatherHourly = pd.concat([weatherHourlyWithoutAbnormal, weatherHourlyWithAbnormal])
```

```python
weatherHourly.head()
```

```python
set(weatherHourly['HourlyStationPressure'])
```

```python
weatherHourly.to_csv('clean_weather_with_abnormal.csv', index=False)
```
